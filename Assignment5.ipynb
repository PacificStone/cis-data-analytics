```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
olympics = pd.read_csv('olympics-final.csv', header=[0,1], index_col=0)
```

# Histogram Using Matplotlib

```python
olympics_hist = plt.hist(olympics[('Total', 'Total Medals')], 
                         range=(0, len(olympics.index) - 1), bins=20, histtype='barstacked')
plt.show()
```

# Histogram Using Seaborn

```python
sea_hist = sns.histplot(data=olympics[('Total', 'Total Medals')], binrange=(0, len(olympics.index) - 1),
            bins=20)
sea_hist.set(xlabel='Total Medals', ylabel='Number of Countries', title='Medal Counts')
plt.show()
```

# Pie Chart

```python
top_countries = olympics.sort_values(by=('Total', 'Total Medals'), ascending=False).head(11)
top_countries.drop(index=top_countries.index[0], axis=0, inplace=True)
my_labels = top_countries.index
total = 100
pie_chart = plt.pie(top_countries[('Total', 'Total Medals')], labels= my_labels, 
                   autopct=lambda p: '{:.0f}%'.format(p * total / 100))
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
plt.show()
```

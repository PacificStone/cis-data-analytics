### Era
This will be categorical and it will be a string

This will be categorical and it will be a bool

### Downloads
This will be numerical and it will be an int


```python
import pandas as pd

df = pd.read_csv('podcasts.csv', encoding='latin1')

df = df.drop(columns=['Unnamed: 8', 'Unnamed: 9'])
df.style.hide_index()
```

```python
range_pods = df.loc[df['Downloads'].between(1000000, 5000000)]
range_pods.style.hide_index()
```

```python
ancient_pods_sorted = df.loc[df['Era'] == 'Ancient']
ancient_pods_sorted = ancient_pods_sorted.sort_values(by=['Downloads'], ascending=False)
ancient_pods_sorted
```

```python
ancient_date = ancient_pods_sorted['Release date'].iloc[0]
print(ancient_date)
```

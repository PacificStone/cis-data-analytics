```python
import pandas as pd
import numpy as np
```

# Initial upload of csv
This is where I first uploaded the csv and printed it to make sure that it was successful

```python
olympics = pd.read_csv('olympics.csv')
olympics
```

# Change Headers
The first thing that I notice is that the columns headers are one row below where they need to be. I will change this by setting new_header to the first row and deleting thetop two rows. Then I will set the columns equal to the first row that I deleted.

```python
new_header = olympics.iloc[0]
olympics = olympics[1:]
olympics.columns = new_header
olympics
```

# Fix Header
Next I notice that the headers are not very descriptive or missing values so I will replace them while turning the columns into a multi-index for greater context. 

```python

multi_tuple = [('Country', 'Name'),
               ('Summer', 'Games Participated'),
               ('Summer', 'Gold'),
               ('Summer', 'Silver'),
               ('Summer', 'Bronze'),
               ('Summer', 'Total'),
               ('Winter', 'Games Participated'),
               ('Winter', 'Gold'),
               ('Winter', 'Silver'),
               ('Winter', 'Bronze'),
               ('Winter', 'Total'),
               ('Total', 'Games Participated'),
               ('Total', 'Gold'),
               ('Total', 'Silver'),
               ('Total', 'Bronze'),
               ('Total', 'Total Medals')]

olympics.columns = pd.MultiIndex.from_tuples(multi_tuple)

olympics
```

# Check Duplicates
I will run a check for duplicates and print a chart of any duplicates that I find

```python
duplicates = olympics[olympics.duplicated()]
duplicates
```

# Re-sort based on Country Name
No duplicates found. Now I notice that the table is sorted by the countries symbol and not there name. For example, Chinese Tapei is found in the T's and the International olympic committee is with the Z's. I will fix this by resorting these columns according to their countries names. 

```python
df1 = olympics[olympics[('Country','Name')]!='Totals'].sort_values(by=('Country','Name'))
df1 = df1.append(olympics[olympics[('Country','Name')]=='Totals'])
olympics = df1
olympics
```

# Fix country abbreviation and create column for it
I noticed that the abbreviations for the countries are clunky and some of them have multiple abbreviations. It seems that the first one is right so I will change it so that there is a seperate column for abbreviations and then make it so it only shows the first abbreviation. 

```python
names_list = olympics[('Country', 'Name')]
names_list = names_list.str.split("(")
names_list = names_list.tolist()
nested_list = names_list[146]
nested_list.append('')

country_names_list = []
countryabbr = []

for i in range(len(names_list)):
    country_names_list.append(names_list[i][0])
    countryabbr.append(names_list[i][1])

olympics[('Country', 'Name')] = country_names_list
olympics[('Country', 'Abbreviation')] = countryabbr

abbr_list = olympics[('Country', 'Abbreviation')]
abbr_list = abbr_list.str.split(")")
abbr_list = abbr_list.tolist()

abbr_final = []

for i in range(len(abbr_list)):
    abbr_final.append(abbr_list[i][0])
               
olympics[('Country', 'Abbreviation')] = abbr_final

# Changing the order of the columns to be clearer
olympics = olympics.sort_index(axis=1, level=0, sort_remaining=False)
olympics = olympics[['Country', 'Summer', 'Winter', 'Total']]
```

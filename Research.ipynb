```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('pokemon_data.csv')
df
```

```python
df[df.duplicated(['Name'], keep=False)]
```

```python
df = df.drop_duplicates(subset=['Name'])
df[df.duplicated(['Name'], keep=False)]
```

```python
grouped = df.groupby(by='Generation').mean().round(2)
sns.relplot(data=grouped, x='Speed', y='Attack', hue='Generation', aspect=1.2, palette='deep')
plt.show()
```

```python
sns.barplot(data=df, x='Sp. Atk', y='Type 1', palette='Spectral', ci=None)
plt.show()
```

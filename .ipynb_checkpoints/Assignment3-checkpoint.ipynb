```python
import pandas as pd 
import numpy as np
```

```python
movies = pd.read_csv('imdb-1.csv')
movies
```

```python
max_rating = movies['Rating'].max()
max_rating
```

```python
early_year = movies['Year'].min()
early_movie = movies[movies['Year']==early_year]['Title']

print(f"The earliest movie in the database is \"{early_movie.iloc[0]}\" which was "
f"made in {earliest_year[0]}")
```

```python
ratings_sort = movies.sort_values(by='Rating', ascending=False)

ratings_sort.head(20)
```

```python
dec_40 = movies[movies['Decade']==1940]
round(dec_40['Year'].mean())
```

```python
ninety_nine = movies.loc[movies['Year']==1999]
ninety_nine
```

```python
really_highly_rated = movies.loc[movies['Rating'] > 8.5]
really_highly_rated
```

```python
before_2000 = movies[movies['Decade'] < 2000]
after_2000 = movies[movies['Decade'] >= 2000]
avg_before_2000 = round(before_2000['Rating'].mean(), 2)
avg_after_2000 = round(after_2000['Rating'].mean(), 2)

print(f"""The average rating of movies made before 2000 is {avg_before_2000},
and the average rating of movies after 2000 is {avg_after_2000}. """)
```

```python
count_before_2000 = before_2000.shape[0]
count_after_2000 = after_2000.shape[0]

proportion_20th = (count_before_2000 / movies.shape[0]) * 100
proportion_21st = (count_after_2000 / movies.shape[0]) * 100

print(f"The proportion of movies in the 20th century is {proportion_20th}%,"
     f"and the proportion of movies in the 21st century is {proportion_21st}%.")
```
